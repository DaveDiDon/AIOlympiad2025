## Numpy
[Documentation](https://numpy.org/doc/stable/index.html)

Numpy, or Numerical Python, is the library upon which most other popular libraries used in machine learning are built. Numpy, most notably, includes the ndarray datatype. This is a multidemensional array object: a gridlike structure that holds data in any number of dimmentions.

Numpy arrays are typically used in place of native python data types.

####Python Lists vs Numpy Arrays

Python Lists

*   Can contain different data types
*   Uses more space in memory than numpy arrays

Numpy Arrays

*   Can only contain a single data type
*   Uses less space in memory than python lists








```python
import numpy as np #Note that as long as the runtime is active I don't need to re-import these, I'm just doing it in case yours resets and you don't realize!

#This is an example of a 1-dementional array. It looks like a python list
lst = [1,3,2]
arr = np.array(lst)
print(arr)
type(arr)

#Here is a two dimentional array, a list of lists
lst2 = [[2,9,3],[5,6,9],[9,2,7]]
arr2 = np.array(lst2)
print(arr2)
type(arr2)
```

```python
import numpy as np

#Instead of creating an array from a python list, you can use some numpy functions
print("Like this:", np.zeros((5,3))) #accepts tuple with array shape
print("Or this:", np.random.random((2,4))) #accepts tuple with array shape
print("Oooor this:", np.arange(-2,3,6)) #Can also accept a third value for step size
```

```python
#One of the coolest things about numpy arrays is that you can make changes to each element of the array at once, for example:

arr = np.array([1, 2, 3, 4, 5])
arr_addone = arr + 1
#print(arr)
#print(arr_addone)

arr_add3 = arr
arr_add3 += 3 #This is the same as saying arr_add3 = arr_add3 + 3
#print(arr)
print(arr_add3)

#This is much faster than using python to loop through the array, and add to each object
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

x = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])

model = LinearRegression()
model.fit(x, y)

print("Prediction for 5:", model.predict([[5]]))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Re-creating the playlist DataFrame with the latest song data
# This part is necessary to ensure the data is available for the visualization
playlist_data = {
    'Song': [
        'Save Your Tears',
        'Every Teardrop Is a Waterfall',
        'Viva la Vida',
        'HandClap',
        'Riptide',
        'Make You Mine',
        'Out of My League',
        'A Sky Full of Stars'
    ],
    'Artist': [
        'The Weeknd',
        'Coldplay',
        'Coldplay',
        'Fitz and The Tantrums',
        'Vance Joy',
        'PUBLIC',
        'Fitz and The Tantrums',
        'Coldplay'
    ],
    'Energy': [
        0.7,
        0.8,
        0.8,
        1.0,
        0.5,
        0.8,
        0.9,
        0.9
    ],
    'Day_Played': [
        'Monday',
        'Tuesday',
        'Wednesday',
        'Thursday',
        'Friday',
        'Saturday',
        'Sunday',
        'Monday'
    ]
}

playlist = pd.DataFrame(playlist_data)

# Define the desired order of days for consistent plotting
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# --- Question 5: Create a visualization for average energy level per day ---
# Calculate the average energy level for each day
average_energy_per_day = playlist.groupby('Day_Played')['Energy'].mean()

# Reindex to ensure all days are present and in the correct order
average_energy_per_day = average_energy_per_day.reindex(days_order, fill_value=0) # Use 0 if a day has no songs

# Create the bar chart for average energy per day
plt.figure(figsize=(10, 6))
plt.bar(average_energy_per_day.index, average_energy_per_day.values, color='lightcoral')

plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Average Energy Level', fontsize=12)
plt.title('Average Energy Level of Music Per Day', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1) # Energy is on a 0-1 scale
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# Print why this visualization is meaningful
print("\n--- Meaningfulness of Average Energy Level Per Day Chart ---")
print("This visualization is meaningful because it allows us to see if there's a pattern in the energy level of the music you listen to throughout the week.")
print("For example, you might notice that you prefer higher energy songs on weekdays to stay motivated, or lower energy songs on weekends for relaxation.")
print("It helps reveal your listening habits and how they might correlate with different days, potentially indicating your mood or activity levels.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np # Used for arange in yticks

# --- Configuration ---
# Define known energy levels for specific songs (can be expanded)
KNOWN_SONG_ENERGY_MAP = {
    'Save Your Tears': 0.7,
    'Every Teardrop Is a Waterfall': 0.8,
    'Viva la Vida': 0.8,
    'HandClap': 1.0,
    'Riptide': 0.5,
    'Make You Mine': 0.8,
    'Out of My League': 0.9,
    'A Sky Full of Stars': 0.9,
    'Blinding Lights': 0.9,
    'The Less I Know The Better': 0.6,
    'Fix You': 0.4,
    'Mr. Brightside': 0.9,
    'Watermelon Sugar': 0.7,
    'Sunflower': 0.8,
    'Thunderstruck': 1.0,
    'Imagine': 0.3,
    'Don\'t Stop Believin\'': 0.9,
    'Lovely Day': 0.6,
    'Africa': 0.7,
    'Sweet Child o\' Mine': 0.9,
    'Again': 0.85 # As discussed for YUI
}

# Define the desired order of days for consistent plotting
DAYS_ORDER = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# --- Hardcoded Playlist Data (Recorded Data) ---
# This data is now directly in the script, no external file needed.
# It's a larger, more varied set to make analysis more meaningful.
recorded_playlist_data = [
    {'Song': 'Save Your Tears', 'Artist': 'The Weeknd', 'Energy': 0.7, 'Day_Played': 'Monday'},
    {'Song': 'Every Teardrop Is a Waterfall', 'Artist': 'Coldplay', 'Energy': 0.8, 'Day_Played': 'Tuesday'},
    {'Song': 'Viva la Vida', 'Artist': 'Coldplay', 'Energy': 0.8, 'Day_Played': 'Wednesday'},
    {'Song': 'HandClap', 'Artist': 'Fitz and The Tantrums', 'Energy': 1.0, 'Day_Played': 'Thursday'},
    {'Song': 'Riptide', 'Artist': 'Vance Joy', 'Energy': 0.5, 'Day_Played': 'Friday'},
    {'Song': 'Make You Mine', 'Artist': 'PUBLIC', 'Energy': 0.8, 'Day_Played': 'Saturday'},
    {'Song': 'Out of My League', 'Artist': 'Fitz and The Tantrums', 'Energy': 0.9, 'Day_Played': 'Sunday'},
    {'Song': 'A Sky Full of Stars', 'Artist': 'Coldplay', 'Energy': 0.9, 'Day_Played': 'Monday'},
    {'Song': 'Blinding Lights', 'Artist': 'The Weeknd', 'Energy': 0.9, 'Day_Played': 'Tuesday'},
    {'Song': 'The Less I Know The Better', 'Artist': 'Tame Impala', 'Energy': 0.6, 'Day_Played': 'Wednesday'},
    {'Song': 'Fix You', 'Artist': 'Coldplay', 'Energy': 0.4, 'Day_Played': 'Thursday'},
    {'Song': 'Mr. Brightside', 'Artist': 'The Killers', 'Energy': 0.9, 'Day_Played': 'Friday'},
    {'Song': 'Watermelon Sugar', 'Artist': 'Harry Styles', 'Energy': 0.7, 'Day_Played': 'Saturday'},
    {'Song': 'Sunflower', 'Artist': 'Post Malone', 'Energy': 0.8, 'Day_Played': 'Sunday'},
    {'Song': 'Thunderstruck', 'Artist': 'AC/DC', 'Energy': 1.0, 'Day_Played': 'Monday'},
    {'Song': 'Imagine', 'Artist': 'John Lennon', 'Energy': 0.3, 'Day_Played': 'Tuesday'},
    {'Song': 'Don\'t Stop Believin\'', 'Artist': 'Journey', 'Energy': 0.9, 'Day_Played': 'Wednesday'},
    {'Song': 'Lovely Day', 'Artist': 'Bill Withers', 'Energy': 0.6, 'Day_Played': 'Thursday'},
    {'Song': 'Africa', 'Artist': 'Toto', 'Energy': 0.7, 'Day_Played': 'Friday'},
    {'Song': 'Sweet Child o\' Mine', 'Artist': 'Guns N\' Roses', 'Energy': 0.9, 'Day_Played': 'Saturday'},
    {'Song': 'Again', 'Artist': 'YUI', 'Energy': 0.85, 'Day_Played': 'Sunday'}, # Added YUI song
    {'Song': 'Bohemian Rhapsody', 'Artist': 'Queen', 'Energy': 0.8, 'Day_Played': 'Monday'},
    {'Song': 'Hotel California', 'Artist': 'Eagles', 'Energy': 0.6, 'Day_Played': 'Tuesday'},
    {'Song': 'Billie Jean', 'Artist': 'Michael Jackson', 'Energy': 0.9, 'Day_Played': 'Wednesday'},
    {'Song': 'Stairway to Heaven', 'Artist': 'Led Zeppelin', 'Energy': 0.5, 'Day_Played': 'Thursday'},
    {'Song': 'Smells Like Teen Spirit', 'Artist': 'Nirvana', 'Energy': 1.0, 'Day_Played': 'Friday'},
    {'Song': 'Yesterday', 'Artist': 'The Beatles', 'Energy': 0.2, 'Day_Played': 'Saturday'},
    {'Song': 'Hallelujah', 'Artist': 'Leonard Cohen', 'Energy': 0.3, 'Day_Played': 'Sunday'},
    {'Song': 'Africa', 'Artist': 'Toto', 'Energy': 0.7, 'Day_Played': 'Monday'}, # Repeat artist
    {'Song': 'Blinding Lights', 'Artist': 'The Weeknd', 'Energy': 0.9, 'Day_Played': 'Tuesday'}, # Repeat artist
    {'Song': 'Viva la Vida', 'Artist': 'Coldplay', 'Energy': 0.8, 'Day_Played': 'Wednesday'}, # Repeat artist
    {'Song': 'Don\'t Stop Believin\'', 'Artist': 'Journey', 'Energy': 0.9, 'Day_Played': 'Thursday'}, # Repeat artist
    {'Song': 'Mr. Brightside', 'Artist': 'The Killers', 'Energy': 0.9, 'Day_Played': 'Friday'}, # Repeat artist
    {'Song': 'Watermelon Sugar', 'Artist': 'Harry Styles', 'Energy': 0.7, 'Day_Played': 'Saturday'}, # Repeat artist
    {'Song': 'Sunflower', 'Artist': 'Post Malone', 'Energy': 0.8, 'Day_Played': 'Sunday'}, # Repeat artist
    {'Song': 'Thunderstruck', 'Artist': 'AC/DC', 'Energy': 1.0, 'Day_Played': 'Monday'}, # Repeat artist
    {'Song': 'Imagine', 'Artist': 'John Lennon', 'Energy': 0.3, 'Day_Played': 'Tuesday'}, # Repeat artist
]


# --- Analysis Functions ---
def analyze_top_artist(df):
    """Evaluates and prints the most listened-to artist."""
    if df.empty:
        print("No data to analyze for top artist.")
        return

    artist_counts = df['Artist'].value_counts()
    if not artist_counts.empty:
        top_artist = artist_counts.index[0]
        top_artist_count = artist_counts.iloc[0]
        print(f"\nYour top (most listened to) artist is: {top_artist} (with {top_artist_count} songs)")
    else:
        print("\nCould not determine top artist (no artists found).")

def analyze_average_energy(df):
    """Calculates and prints the average energy level with a descriptive message."""
    if df.empty:
        print("No data to analyze for average energy level.")
        return

    average_energy = df['Energy'].mean()
    print(f"\nYour playlist's average energy level is: {average_energy:.2f}")

    if average_energy < 0.4:
        print("I'm just a chill guy. Your playlist leans towards relaxed vibes.")
    elif average_energy > 0.6:
        print("Looks like you're ready to power through anything! Your playlist is packed with high-energy tracks.")
    else: # Between 0.4 and 0.6 inclusive
        print("You've got a balanced mix of vibes going on! Your playlist has a good blend of energy levels.")

# --- Visualization Functions ---
def plot_music_played_per_day(df):
    """
    Creates a bar chart showing the number of songs played per day.
    """
    if df.empty:
        print("No data to plot music played per day.")
        return

    day_counts = df['Day_Played'].value_counts().reindex(DAYS_ORDER, fill_value=0)

    plt.figure(figsize=(10, 6))
    plt.bar(day_counts.index, day_counts.values, color='#60a5fa', edgecolor='black') # Blue-400
    plt.xlabel('Day of the Week', fontsize=12)
    plt.ylabel('Number of Songs Played', fontsize=12)
    plt.title('Music Played Per Day', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(np.arange(0, day_counts.max() + 2, 1), fontsize=10) # Ensure integer ticks
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    most_played_day = day_counts.idxmax()
    max_plays = day_counts.max()
    print(f"\nI think I listen to music mostly on {most_played_day} because that's when the most songs ({max_plays}) were played in this playlist.")
    print("Meaningfulness: This chart clearly shows your listening frequency across different days, highlighting peak listening days. It can reveal patterns in when you engage with music the most.")

def plot_average_energy_per_day(df):
    """
    Creates a bar chart showing the average energy level of music per day.
    """
    if df.empty:
        print("No data to plot average energy per day.")
        return

    # Calculate average energy for each day
    average_energy_per_day = df.groupby('Day_Played')['Energy'].mean().reindex(DAYS_ORDER, fill_value=0)

    plt.figure(figsize=(10, 6))
    plt.bar(average_energy_per_day.index, average_energy_per_day.values, color='#f87171', edgecolor='black') # Red-400
    plt.xlabel('Day of the Week', fontsize=12)
    plt.ylabel('Average Energy Level (0-1)', fontsize=12)
    plt.title('Average Energy Level of Music Per Day', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.ylim(0, 1) # Energy is on a 0-1 scale
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    print("\nMeaningfulness: This visualization helps you understand if your music energy preferences change throughout the week.")
    print("For instance, you might see higher energy levels on workout days or lower energy levels on relaxation days, revealing a correlation between your daily activities/moods and music choice. It offers insights into your weekly emotional or activity rhythm through music.")

def plot_energy_distribution(df):
    """
    Creates a histogram showing the distribution of energy levels across the entire playlist.
    """
    if df.empty:
        print("No data to plot energy distribution.")
        return

    plt.figure(figsize=(10, 6))
    # Using 10 bins for the 0-1 range, so each bin is 0.1 wide
    plt.hist(df['Energy'], bins=np.arange(0, 1.1, 0.1), color='#8a2be2', edgecolor='black', alpha=0.7) # BlueViolet
    plt.xlabel('Energy Level (0-1)', fontsize=12)
    plt.ylabel('Number of Songs', fontsize=12)
    plt.title('Distribution of Energy Levels in Playlist', fontsize=14, fontweight='bold')
    plt.xticks(np.arange(0, 1.1, 0.1), fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    print("\nMeaningfulness: This histogram provides an overview of the overall energy profile of your playlist.")
    print("It shows you how many songs fall into low, medium, or high energy categories. For example, a peak in the 0.8-0.9 range indicates a preference for high-energy tracks, while a peak around 0.3-0.4 suggests a preference for more relaxed music. This helps you understand the general 'vibe' of your entire collection.")


# --- Main Application Flow ---
if __name__ == "__main__":
    print("--- Running Playlist Analyzer (Self-Contained Version) ---")

    # Create DataFrame directly from hardcoded data
    playlist_df = pd.DataFrame(recorded_playlist_data)

    if not playlist_df.empty:
        # 1. Perform analyses
        analyze_top_artist(playlist_df)
        analyze_average_energy(playlist_df)

        # 2. Generate visualizations
        plot_music_played_per_day(playlist_df)
        plot_average_energy_per_day(playlist_df)
        plot_energy_distribution(playlist_df) # New visualization
    else:
        print("\nNo playlist data available in the script. Please check 'recorded_playlist_data'.")
```
